### 1.1.1
Start by plotting the capacity factors for wind and solar throughout the first two weeks in January
and the first two weeks in July. Do this for the most recent year for which you have available data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read csv files for PV and onshore wind capacity factors
pv = pd.read_csv("pv_optimal.csv", sep=";")
wind = pd.read_csv("onshore_wind_1979-2017.csv", sep=";")

# Set time as index and convert to datetime
pv["utc_time"] = pd.to_datetime(pv["utc_time"])
wind["utc_time"] = pd.to_datetime(wind["utc_time"])

# Lock PV capacity factors only from january 2017
pv_jan = pv[(pv["utc_time"].dt.month == 1) & (pv["utc_time"].dt.year == 2017)]
# Lock PV capacity factors only from june 2017
pv_june = pv[(pv["utc_time"].dt.month == 6) & (pv["utc_time"].dt.year == 2017)]

# Lock wind capacity factors only from january 2017
wind_jan = wind[(wind["utc_time"].dt.month == 1) & (wind["utc_time"].dt.year == 2017)]
# Lock wind capacity factors only from june 2017
wind_june = wind[(wind["utc_time"].dt.month == 6) & (wind["utc_time"].dt.year == 2017)]

# Plot PV data for DK in january
plt.figure(figsize=(10, 5))
plt.bar(pv_jan["utc_time"], pv_jan["DNK"], label="PV DK - January")
plt.bar(pv_june["utc_time"], pv_june["DNK"], label="PV DK - June")
plt.xlabel("Time")
plt.ylabel("Capacity Factor for PV")
plt.title("Capacity Factor for PV in Denmark (January vs June)")
plt.legend()
plt.show()


# Plot wind data for DK in january
plt.figure(figsize=(10, 5))
plt.bar(wind_jan["utc_time"], wind_jan["DNK"], label="Wind DK - January")
plt.bar(wind_june["utc_time"], wind_june["DNK"], label="Wind DK - June")
plt.xlabel("Time")
plt.ylabel("Capacity Factor for Wind")
plt.title("Capacity Factor for Wind in Denmark (January vs June)")
plt.legend()
plt.show()

```
